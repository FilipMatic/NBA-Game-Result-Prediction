```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
#Importing Dataset
X = pd.read_csv("nba.games.stats.csv")
X.head(10)
```

```python
#Intitally done when we were going to use Home and Away as a category but decided to focus on the statistics within the game  redundant in final feature selection


one_hot = pd.get_dummies(X['Home'],prefix='Side')
X = X.drop('Home',axis = 1)
X = X.join(one_hot)
X.head()
```

```python
#changing team labels to unique numbers
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X['Team']=le.fit_transform(X['Team'])
X['Opponent']=le.transform(X['Opponent'])
X.head()
```

```python
#mapping a win or loss to 1=Win and Zero= loss, also feature selection is done here. 


d=X['WINorLOSS']
d=list(map(lambda x: 1 if x != 'L' else 0, d))

featuretoDrop = ['Unnamed: 0', 'Game', 'Date',  'WINorLOSS','TeamPoints',
                 'FieldGoals', 'FieldGoalsAttempted', 'FieldGoals.', 'X3PointShots', 
                 'X3PointShotsAttempted', 'X3PointShots.', 'FreeThrows', 'FreeThrowsAttempted', 
                 'FreeThrows.', 'OffRebounds', 'TotalRebounds', 'Assists','Turnovers','Opp.TotalRebounds', 'Opp.Steals', 'Opp.Blocks', 'Side_Away', 'Side_Home', 'Team', 'Opponent', 'Opp.TotalFouls']
data = X.drop(featuretoDrop, axis = 1)
data.head(15)
features=list(data)
print(features)
```

```python
from sklearn.model_selection import train_test_split
#splitting data set into training and test set
X_train, X_test, WL_train, WL_test = train_test_split(data[features], d, test_size=0.33, random_state=42)

X_train = X_train.to_numpy()
X_test = X_test.to_numpy()
```

```python
print(X_train)
```

```python
from sklearn import svm


#training the model 
clf_svm = svm.SVC(kernel="linear", random_state=0, max_iter=3500).fit(X_train, WL_train)
```

```python
predicted_svm = clf_svm.predict(X_test)#prediction of test set
#printing accuracy of predictions
correct = 0
for tag, pred in zip(WL_train, predicted_svm):   # zip allows to go through two lists simultaneously
    if (tag == pred):
        correct += 1
accuracy = correct/len(WL_test)

print(accuracy)
```
